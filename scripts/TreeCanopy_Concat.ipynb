```python
import pandas as pd
import geopandas as gpd
```

```python
# Read raw file
```

```python
gdf1 = gpd.read_file("C://Users//zlfdo//Desktop//ECCE GeoTrio//App2_Potential//Data//TreeComm//Tree_Comm_2012.shp")
gdf2 = gpd.read_file("C://Users//zlfdo//Desktop//ECCE GeoTrio//App2_Potential//Data//TreeComm//Tree_Comm_2013.shp")
gdf3 = gpd.read_file("C://Users//zlfdo//Desktop//ECCE GeoTrio//App2_Potential//Data//TreeComm//Tree_Comm_2015.shp")
gdf4 = gpd.read_file("C://Users//zlfdo//Desktop//ECCE GeoTrio//App2_Potential//Data//TreeComm//Tree_Comm_2017.shp")
gdf5 = gpd.read_file("C://Users//zlfdo//Desktop//ECCE GeoTrio//App2_Potential//Data//TreeComm//Tree_Comm_2020.shp")
```

```python
gdf1.head()
```

```python
# Convert to dataframe
```

```python
df1 = gdf1
df2 = gdf2
df3 = gdf3
df4 = gdf4
df5 = gdf5
```

```python
# Concat
```

```python
frames = [df1, df2, df3, df4, df5]
result = pd.concat(frames)
result.head()
```

```python
# Convert back to geodataframe
```

```python
gdf = gpd.GeoDataFrame(result, geometry=result['geometry'])
gdf.head()
```

```python
# Output
```

```python
gdf.to_file('TC_concat.shp')
```
